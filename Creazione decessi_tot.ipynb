```python
import pandas as pd
df = pd.read_csv('csv/decessi.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
#elimino i valori 0 e n.d. della colonna Totale per sfoltire df
df=df.set_index('Totale')
df=df.drop('0', axis=0)
df=df.drop('n.d.', axis=0)
df=df.reset_index()
```

```python
df['Data'] = pd.to_datetime(df['Data'], infer_datetime_format=True)
df.head()
```

```python
 df['Totale'] = pd.to_numeric(df['Totale'])
```

```python
#elimino le colonne che al momento non ci interessano
df.drop(['REG', 'PROV', 'NOME_REGIONE', 'NOME_PROVINCIA', 'NOME_COMUNE', 'COD_PROVCOM', 'TIPO_COMUNE',
                 'CL_ETA', 'GE', 'Maschi', 'Femmine'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#somme i decessi in base alla data
df=df.groupby('Data').sum()
df
```

```python
df.to_csv('csv/decessitot.csv')
```

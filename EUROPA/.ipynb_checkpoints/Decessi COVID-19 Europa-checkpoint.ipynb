<h1>CREAZIONE SERIE STORICHE DEI DECESSI PER COVID-19 DELLE NAZIONI EUROPEE

```python
import pandas as pd
import numpy as np

df = pd.read_csv('csv/decessi_covid19_eu.csv')
df.head()
```

```python
df.rename(columns={'CountryName': 'Geo'}, inplace=True)
df.rename(columns={'CumulativeDeceased': 'Value'}, inplace=True)
df.drop(['iso3', 'lat', 'lon', 'CumulativePositive', 'CumulativeRecovered', 'CurrentlyPositive', 'Hospitalized',
        'IntensiveCare', 'EUcountry', 'EUCPMcountry', 'NUTS'], axis=1, inplace=True)
df.head()
```

```python
df['Date']= pd.to_datetime(df['Date'])
df['week']= df['Date'].dt.week
df['Date'] = df['Date'].astype(str) + 'W' + df['week'].astype(str)
df.head()
```

```python
df.drop(['week'], axis=1, inplace=True)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'] + '1', format='%Y-%m-%dW%W%w')
df.head()
```

<h1>FRANCIA

```python
df_fr = df[df['Geo'] == 'France']
df_fr
```

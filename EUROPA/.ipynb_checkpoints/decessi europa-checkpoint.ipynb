```python
import pandas as pd
df = pd.read_csv('decessi_eu_week.csv')
df.head()
```

```python
df.drop(['SEX', 'UNIT'], axis=1, inplace=True)
df.head()
```

```python
df.replace({":": 0}, inplace=True)
df['Value'] = pd.to_numeric(df['Value'])
df.head()
```

```python
df['TIME'] = pd.to_datetime(df['TIME'] + '0', format='%YW%W%w')
df.head()
```

```python
df_fr = df[df['GEO'] == 'France']
df_fr.head()
```

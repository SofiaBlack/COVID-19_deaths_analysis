<h2>CREAZIONE SERIE STORICHE DEI DECESSI COVID ACCERATI SU BASE REGIONALE IN BASE ALLA PERCENTUALE DELLA COPERTURA COMUNALE DEI DECESSI TOTALE

```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
df = pd.read_csv('csv/decessi_covid19_regioni.csv')
df.head()
```

```python
df.drop(['stato', 'codice_regione', 'lat', 'long', 'ricoverati_con_sintomi','terapia_intensiva', 'totale_ospedalizzati',
        'isolamento_domiciliare', 'variazione_totale_positivi', 'nuovi_positivi', 'dimessi_guariti', 
         'casi_da_sospetto_diagnostico', 'casi_da_screening','totale_casi','tamponi', 'casi_testati', 'note', 
         'totale_positivi'], axis=1, inplace=True)
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'], format='%Y-%m-%d')
df['data'] = df['data'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'], format='%Y-%m-%d')
df['deceduti'] = pd.to_numeric(df['deceduti'])
```

Estrapolazione dei decessi COVID-19 accertati relativi ai mesi di <b>marzo</b>, <b>aprile</b>, <b>maggio</b>, <b>giugno</b>, <b>luglio</b>, <b>agosto</b> e <b>settembre</b>

```python
df = df[df['data'] > '2020-02-29']
df = df[df['data'] < '2020-10-01']
df
```

```python
#numero di record
index = df.index
record = len(index)
print(record)
```

```python
import numpy as np
deceduti = df['deceduti']
print('mean', np.mean(deceduti))
print('min', np.min(deceduti))
print('max', np.max(deceduti))
```

Dai dati diffusi dalla Protezione Civile italiana sono stati estratti i decessi sudivisi per ogni regione. 
<br />
Per ottenere il numero totale di decessi COVID-19 per ogni regione nei mesi di marzo, aprile, maggio e giugno; la colonna `decessi` contenente il valore incrementanto dei decessi Ã¨ stata sottoposta all'operazione di differenza. 
<br />
Successivamente i valori giornalieri sono stati raggruppati a livello mensile.
<br />
La serie storica ottenuta Ã¨ stata a sua volta moltiplicata per la percentuale di copertura comunale relativa ai decessi totali ISTAT con cui saranno confrontate le serie storiche regionali dei decessi COVID-19 ottenute.

<h2>ABRUZZO</h2>

```python
abruzzo = df[df['denominazione_regione'] =='Abruzzo']
abruzzo.head()
```

```python
abruzzo.drop(['denominazione_regione'], axis=1, inplace=True)
abruzzo.head()
```

```python
abruzzo=abruzzo.set_index('data')
abruzzo.head()
```

```python
abruzzo = abruzzo.diff()
abruzzo.fillna(0, inplace=True)
abruzzo.head()
```

```python
abruzzo = abruzzo.groupby(pd.Grouper(freq='M')).sum()
abruzzo
```

```python
abruzzo.to_csv('csv/regioni_covid/abruzzo.csv')
```

<h2>CREAZIONE SERIE STORICHE DEI DECESSI TOTALI 
    MENSILI REGIONALI DAL 2015 A SETTEMBRE 2020</h2>

```python
import matplotlib.pyplot as plt

import pandas as pd
df = pd.read_csv('csv/serieStorica_30settembre2020.csv')
df.head()
```

```python
df['DATA'] = pd.to_datetime(df['DATA'], format='%Y-%m-%d')
```

```python
df.drop(['REG', 'PROV', 'NOME_PROVINCIA', 'COD_PROVCOM', 'NOME_COMUNE'], axis=1, inplace=True)
df.head()
```

<h2>ABRUZZO</h2>

```python
abruzzo = df[df['NOME_REGIONE']=='Abruzzo']
abruzzo.head()
```

```python
abruzzo=abruzzo.groupby('DATA').sum()
abruzzo
```

```python
abruzzo = abruzzo.groupby(pd.Grouper(freq='M')).sum()
abruzzo
```

```python
plt.figure(figsize=(12,6))
plt.plot(abruzzo)
plt.show()
```

```python
abruzzo.to_csv('csv/regioni/abruzzo.csv')
```
